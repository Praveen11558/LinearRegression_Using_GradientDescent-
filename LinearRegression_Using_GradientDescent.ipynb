```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Generate random inputs data to train on

```python
observations= 1000

xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(-10,10,size=(observations,1))

inputs = np.column_stack((xs,zs))

print(inputs)
inputs.shape
```

#  Create the target we will aim at

```python
noise= np.random.uniform(-1,-1,size=(observations,1))
print(noise)
targets = 2*xs - 3*zs + 5+ noise

print(targets.shape)
```

```python
print(targets)
```

# Plot the training Data

```python
targets= targets.reshape(observations,)
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(observations,1)
```

# initialise the variables

```python
init_range = 0.1
weights = np.random.uniform(-init_range,init_range, size=(2,1))

baises = np.random.uniform(-init_range,init_range,size=1)
print(weights)
print(baises)
```

# Set a learning rate

```python
learning_rate = 0.01
```

# Train the model

```python
for i in range(200):
    output=np.dot(inputs,weights) + baises
    delta=output-targets
    
    loss = np.sum(delta**2) / 2 / observations
    
    print(loss)
    delta_scaled=delta/observations
    weights = weights -learning_rate*np.dot(inputs.T,delta_scaled)
    baises= baises- learning_rate*np.sum(delta_scaled)

```
